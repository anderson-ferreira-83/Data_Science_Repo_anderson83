```python
# Mapeamento das alternativas para valores numéricos
alternatives_map = {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4}
```

```python
# Convertemos as alternativas para valores numéricos, onde '-1' representa a linha faltante
data_numerical = [[alternatives_map.get(x, -1) for x in row] for row in data]

# Convertendo para DataFrame para facilitar a manipulação
df = pd.DataFrame(data_numerical)
df.columns = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5']
#
df.head(10)
```

```python
df.info()
```

```python
# Inicializando a coluna de rótulos
labels = []

# Percorrendo cada linha do DataFrame
for _, row in df.iterrows():
    # Verificar se existe valor -1 na linha
    if -1 in row.values:
        labels.append(-1)
    else:
        # Contagem de ocorrências de cada elemento
        counts = pd.Series(row).value_counts()
        # Caso 1: Nenhuma repetição
        if all(counts == 1):
            labels.append('v_11111')
        # Caso 2: Uma repetição com frequência 2
        elif (counts == 2).sum() == 1 and (counts == 1).sum() == 3:
            labels.append('v_01112')
        # Caso 5: Uma repetição com frequência 2 e outra com frequência 3 (AJUSTADO)
        elif (counts == 3).sum() == 1 and (counts == 2).sum() == 1:
            labels.append('v_00023')
        # Caso 3: Duas repetições com frequência 2
        elif (counts == 2).sum() == 2:
            labels.append('v_00122')
        # Caso 4: Uma repetição com frequência 3
        elif (counts == 3).sum() == 1:
            labels.append('v_01113')
        # Caso 6: Uma repetição com frequência 4
        elif (counts == 4).sum() == 1:
            labels.append('v_00014')
        # Nenhuma regra aplicada
        else:
            labels.append(-1)

# Adicionando os rótulos ao DataFrame
df['label'] = labels
# Exibindo o DataFrame atualizado com os rótulos
df.head(10)
```

```python
df['label'].unique()
```

```python
# Checagem de variação
df['variation'] = (df['label'] != df['label'].shift()).astype(int)
# O primeiro elemento será 0, pois não existe comparação
df['variation'].iloc[0] = 0
# Exibindo o DataFrame atualizado
df.head(20)
```

## Treinamento dos Modelos

### Divisao do conjunto de dados

```python
# Divisão em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

### Modelo KNN

```python
# Criando o modelo de KNN
k = 3
knn =  KNeighborsClassifier(k)
```

```python
# Treinando o modelo KNN
knn.fit(X_train, y_train)
```

### Modelo Arvore de Decisao

```python
# Criando o modelo de Árvore de Decisão
clf = DecisionTreeClassifier(random_state=42)
```

```python
# Treinando o modelo Arvore de Decisao
clf.fit(X_train, y_train)
```

```python
# Fazendo previsões
y_pred = clf.predict(X_test)
```

```python
# Calculando a acurácia
accuracy = accuracy_score(y_test, y_pred)
print(f"Acurácia: {accuracy:.2f}")
```

```python
# Gerando a matriz de confusão
conf_matrix = confusion_matrix(y_test, y_pred)

# Obtendo os rótulos únicos a partir de y_test e y_pred
labels = np.unique(np.concatenate((y_test, y_pred)))

# Exibindo a matriz de confusão
print("Matriz de Confusão:")
print(conf_matrix)
```

```python
# Relatório de classificação
print("Relatório de Classificação:")
print(classification_report(y_test, y_pred))
```
