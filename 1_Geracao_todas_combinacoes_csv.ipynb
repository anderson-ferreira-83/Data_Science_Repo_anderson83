## 1_11111

```python
import csv
import pandas as pd
import itertools
```

```python
# Letras disponíveis
letras = ['A', 'B', 'C', 'D', 'E']

# Usar um set para armazenar combinações válidas (para evitar duplicatas)
resultado_final = set()

# Gerar todas as permutações das letras
permutacoes = itertools.permutations(letras)

# Adiciona as permutações ao resultado final
resultado_final.update(permutacoes)

# Exibir o número total de combinações e algumas combinações resultantes
print(f"Total de combinações: {len(resultado_final)}")
#for comb in resultado_final:
    #print(''.join(comb))

# Criar uma lista de combinações formatadas
combinacoes_formatadas = [''.join(comb) for comb in resultado_final]

# Criar um DataFrame a partir da lista de combinações
df1 = pd.DataFrame(combinacoes_formatadas, columns=['Combinacoes'])

# Exportar o DataFrame para um arquivo .csv sem cabeçalho
df1.to_csv('1_combinacoes_11111.csv', index=False, header =False)

print("As combinações foram exportadas para 1_combinacoes_11111.csv.")
```

## 2_01112

```python
# Letras disponíveis
letras = ['A', 'B', 'C', 'D', 'E']

# Usar um set para armazenar combinações válidas (para evitar duplicatas)
resultado_final = set()

# Para cada letra, escolhemos uma para ser repetida
for letra_repetida in letras:
    # Escolher 3 letras distintas das 4 restantes
    letras_restantes = [letra for letra in letras if letra != letra_repetida]

    # Gerar combinações de 3 letras distintas das letras restantes
    for combinacao in itertools.combinations(letras_restantes, 3):
        # Formar a lista de 5 letras (2 repetidas + 3 distintas)
        cinco_letras = list(combinacao) + [letra_repetida, letra_repetida]

        # Gerar todas as permutações dessas 5 letras
        permutacoes = set(itertools.permutations(cinco_letras))

        # Adiciona as permutações ao resultado final
        resultado_final.update(permutacoes)

# Exibir o número total de combinações e algumas combinações resultantes
print(f"Total de combinações: {len(resultado_final)}")
#for comb in resultado_final:
   #print(''.join(comb))

# Criar uma lista de combinações formatadas
combinacoes_formatadas = [''.join(comb) for comb in resultado_final]

# Criar um DataFrame a partir da lista de combinações
df2 = pd.DataFrame(combinacoes_formatadas, columns=['Combinacoes'])

# Exportar o DataFrame para um arquivo .csv sem cabeçalho
df2.to_csv('2_combinacoes_01112.csv', index=False, header =False)

print("As combinações foram exportadas para 2_combinacoes_01112.csv.")
```

## 3_00122

```python
# Letras disponíveis
letras = ['A', 'B', 'C', 'D', 'E']

# Usar um set para armazenar combinações válidas (para evitar duplicatas)
resultado_final = set()

# Para cada combinação de 2 letras (para serem repetidas)
for letra1, letra2 in itertools.combinations(letras, 2):
    # Escolher 1 letra diferente das restantes
    letras_restantes = [letra for letra in letras if letra not in [letra1, letra2]]

    for letra_diferente in letras_restantes:
        # Criar a lista de letras (2 de letra1, 2 de letra2, 1 de letra_diferente)
        cinco_letras = [letra1, letra1, letra2, letra2, letra_diferente]

        # Gerar todas as permutações dessas 5 letras
        permutacoes = set(itertools.permutations(cinco_letras))

        # Adiciona as permutações ao resultado final
        resultado_final.update(permutacoes)

# Exibir o número total de combinações
print(f"Total de combinações: {len(resultado_final)}")

# Criar uma lista de combinações formatadas
combinacoes_formatadas = [''.join(comb) for comb in resultado_final]

# Criar um DataFrame a partir da lista de combinações
df3 = pd.DataFrame(combinacoes_formatadas, columns=['Combinacoes'])

# Exportar o DataFrame para um arquivo .csv sem cabeçalho
df3.to_csv('3_combinacoes_00122.csv', index=False, header=False)

print("As combinações foram exportadas para 3_combinacoes_00122.csv.")

```

## 4_00023

```python
# Letras disponíveis
letras = ['A', 'B', 'C', 'D', 'E']

# Usar um set para armazenar combinações válidas (para evitar duplicatas)
resultado_final = set()

# Para cada letra, escolhemos uma para ser repetida 3 vezes
for letra_3x in letras:
    # Escolher uma letra diferente para ser repetida 2 vezes
    letras_restantes = [letra for letra in letras if letra != letra_3x]

    for letra_2x in letras_restantes:
        # Criar a lista de letras (3 de letra_3x e 2 de letra_2x)
        cinco_letras = [letra_3x] * 3 + [letra_2x] * 2

        # Gerar todas as permutações dessas 5 letras
        permutacoes = set(itertools.permutations(cinco_letras))

        # Adiciona as permutações ao resultado final
        resultado_final.update(permutacoes)

# Exibir o número total de combinações
print(f"Total de combinações: {len(resultado_final)}")

# Criar uma lista de combinações formatadas a partir do resultado final
combinacoes_formatadas = [''.join(comb) for comb in resultado_final]

# Criar um DataFrame a partir da lista de combinações
df4 = pd.DataFrame(combinacoes_formatadas, columns=['Combinacoes'])

# Exportar o DataFrame para um arquivo .csv sem cabeçalho
df4.to_csv('4_combinacoes_00023.csv', index=False, header=False)

print("As combinações foram exportadas para 4_combinacoes_00023.csv.")
```

## 5_00014

```python
# Letras disponíveis
letras = ['A', 'B', 'C', 'D', 'E']

# Usar um set para armazenar combinações válidas (para evitar duplicatas)
resultado_final = set()

# Para cada letra, escolhemos uma para ser repetida 4 vezes
for letra_4x in letras:
    # Escolher uma letra diferente para ser usada 1 vez
    letras_restantes = [letra for letra in letras if letra != letra_4x]

    for letra_1x in letras_restantes:
        # Criar a lista de letras (4 de letra_4x e 1 de letra_1x)
        cinco_letras = [letra_4x] * 4 + [letra_1x]

        # Gerar todas as permutações dessas 5 letras
        permutacoes = set(itertools.permutations(cinco_letras))

        # Adiciona as permutações ao resultado final
        resultado_final.update(permutacoes)

# Exibir o número total de combinações
print(f"Total de combinações: {len(resultado_final)}")

# Criar uma lista de combinações formatadas a partir do resultado final
combinacoes_formatadas = [''.join(comb) for comb in resultado_final]

# Criar um DataFrame a partir da lista de combinações
df5 = pd.DataFrame(combinacoes_formatadas, columns=['Combinacoes'])

# Exportar o DataFrame para um arquivo .csv sem cabeçalho
df5.to_csv('5_combinacoes_00014.csv', index=False, header=False)

print("As combinações foram exportadas para 5_combinacoes_00014.csv.")
```

## 6_00113

```python
# Letras disponíveis
letras = ['A', 'B', 'C', 'D', 'E']

# Usar um set para armazenar combinações válidas (para evitar duplicatas)
resultado_final = set()

# Para cada letra, escolhemos uma para ser repetida 3 vezes
for letra_3x in letras:
    # Escolher 2 letras diferentes para serem usadas 1 vez
    letras_restantes = [letra for letra in letras if letra != letra_3x]

    for letra_1x_1 in letras_restantes:
        for letra_1x_2 in letras_restantes:
            if letra_1x_1 != letra_1x_2:  # Garantir que as letras são diferentes
                # Criar a lista de letras (3 de letra_3x e 1 de cada letra_1x)
                cinco_letras = [letra_3x] * 3 + [letra_1x_1, letra_1x_2]

                # Gerar todas as permutações dessas 5 letras
                permutacoes = set(itertools.permutations(cinco_letras))

                # Adiciona as permutações ao resultado final
                resultado_final.update(permutacoes)

# Exibir o número total de combinações
print(f"Total de combinações: {len(resultado_final)}")

# Criar uma lista de combinações formatadas a partir do resultado final
combinacoes_formatadas = [''.join(comb) for comb in resultado_final]

# Criar um DataFrame a partir da lista de combinações
df6 = pd.DataFrame(combinacoes_formatadas, columns=['Combinacoes'])

# Exportar o DataFrame para um arquivo .csv sem cabeçalho
df6.to_csv('6_combinacoes_00113.csv', index=False, header=False)

print("As combinações foram exportadas para 6_combinacoes_00113.csv.")
```
